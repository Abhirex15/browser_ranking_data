```python
import pandas as pd
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import re
import math
```

```python
df = pd.read_csv("/content/browser_rankings_data.csv")
```

```python
X = df['Unnamed: 8']
Y = df['Unnamed: 9']
```

```python
 re_map = {r'\be[\-& ]learning\b': 'elearning'}
 for r, map in re_map.items():
     d = df['Unnamed: 8']
     for i,e in enumerate(d):
         d[i] = re.sub(r, map, e)
```

```python
df["X"] = df["Unnamed: 8"].fillna("")
df["Y"] = df["Unnamed: 9"].fillna("")
```

```python
X = df['X']
Y = df['Y']
```

```python
from nltk.tokenize import sent_tokenize
import nltk
nltk.download('punkt')
```

```python
gen_docs_X = [[w.lower() for w in word_tokenize(text)] 
            for text in X]

        
```

```python
import gensim
dictionary_X = gensim.corpora.Dictionary(gen_docs_X)

#print(dictionary.token2id)
```

```python
corpus_X = [dictionary_X.doc2bow(gen_doc) for gen_doc in gen_docs_X]
```

```python
import numpy as np
tf_idf_X = gensim.models.TfidfModel(corpus_X)
for doc in tf_idf_X[corpus_X]:
  print([[dictionary_X[id], np.around(freq, decimals=2)] for id, freq in doc])
```

```python
sims_X = gensim.similarities.Similarity('/content',tf_idf_X[corpus_X],
                                        num_features=len(dictionary_X))
```

```python

df['Y'] = df['Y'].apply(str)

#df['Y'] = df['Y'].apply(str)
for line in df['Y']:
    query_doc = [w.lower() for w in word_tokenize(line)]
    query_doc_bow = dictionary_X.doc2bow(query_doc) 
```

```python
query_doc_tf_idf = tf_idf_X[query_doc_bow]
# print(document_number, document_similarity)
print('Comparing Result:', *sims_X[query_doc_tf_idf], sep = "\n") 
```
